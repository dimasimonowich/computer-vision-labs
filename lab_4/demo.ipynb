```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from models.photograph import PhotoGraph
```

```python
image_01 = cv.imread("data/image_01.png")
image_02 = cv.imread("data/image_02.png")
image_03 = cv.imread("data/image_03.png")
image_04 = cv.imread("data/image_04.png")
image_05 = cv.imread("data/image_05.png")

mask_01 = cv.imread("data/mask_01.png")[:,:,1]
mask_02 = cv.imread("data/mask_02.png")[:,:,1]
mask_03 = cv.imread("data/mask_03.png")[:,:,1]\
mask_04 = cv.imread("data/mask_04.png")[:,:,1]
mask_05 = cv.imread("data/mask_05.png")[:,:,1]

images = np.array([image_01, image_02, image_03, image_04, image_05])
masks = np.array([mask_01, mask_02, mask_03, mask_04, mask_05])
```

```python
from time import time

start = time()
pg = PhotoGraph(images/255, masks/255, alpha=5, beta=1)
merged_image = pg.merge_images()
end = time()

print(f"Runtime: {end-start}")

plt.imshow(merged_image)
plt.show()
```
